# Task 1: Exploring and Visualizing Iris Dataset

## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries installed successfully!")
```

## Loading the Iris Dataset

```python
import seaborn as sns

iris = sns.load_dataset("iris")

print("Dataset Shape:")
print(iris.shape)

print("\nColumns:")
print(iris.columns)

print("\nFirst 5 Rows:")
print(iris.head())
```

## Scatter Plot Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(
    x="sepal_length",
    y="sepal_width",
    hue="species",
    data=iris
)

plt.title("Sepal Length vs Sepal Width")
plt.show()
```

### Observation

The scatter plot shows the relationship between sepal length and sepal width for different flower species.

## Histogram Analysis

```python
plt.figure(figsize=(8,5))

sns.histplot(iris["sepal_length"], bins=10)

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")

plt.show()
```

### Observation

The histogram shows the distribution of sepal length values in the Iris dataset.

## Box Plot Analysis

```python
plt.figure(figsize=(8,5))

sns.boxplot(y=iris["sepal_length"])

plt.title("Box Plot of Sepal Length")

plt.show()
```

### Observation

The box plot shows the spread of sepal length values and helps identify outliers.

## Conclusion

The Iris dataset contains 150 records and 5 columns. Different visualizations were used to understand the distribution and relationships between features. The analysis helps distinguish flower species based on their measurements.
